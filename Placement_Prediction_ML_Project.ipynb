# Outlier Removal

```python
for i in num:
    plt.figure()
    sns.boxplot(y=i,data=df)
```

```python
for i in num:
    q1=df[i].quantile(0.25)
    q3=df[i].quantile(0.75)
    iqr=q3-q1
    UL=q3+1.5*iqr
    LL=q1-1.5*iqr
    
    for j in df[i]:
        if j>UL:
            df[i]=df[i].replace(i,UL)
        elif j<LL:
            df[i]=df[i].replace(i,LL)
```

```python
for i in num:
    plt.figure()
    sns.boxplot(y=i,data=df)
```

# Encoding

```python
#Encode the cat column
df_new=pd.get_dummies(df,columns=['gender',
  'ssc_b',
  'hsc_b',
  'hsc_s',
  'degree_t',
  'workex',
  'specialisation'])
```

```python
df_new
```

```python
df_new['status'].replace({'Placed':1,'Not Placed':0},inplace=True)
```

```python
df_new
```

```python
x=df_new.drop(['salary','status'],axis=1)
y=df_new['salary']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=20)
```

# Linear Regression Model

```python
from sklearn import linear_model,metrics
```

```python
lr=linear_model.LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```

# Logistic Regression Model

```python
xl=df_new.drop(['salary','status'],axis=1)
yl=df_new['status']
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.model_selection import train_test_split
xl_train,xl_test,yl_train,yl_test=train_test_split(xl,yl,test_size=0.25,random_state=20)
```

```python
LogR=LogisticRegression(solver='liblinear')
```

```python
yl
```

```python
LogR.fit(xl_train,yl_train)
```

```python
yl_pred=LogR.predict(xl_test)
yl_pred
```

```python
r2_score(yl_test,yl_pred)
```

```python
sns.heatmap(metrics.confusion_matrix(yl_pred,yl_test),annot=True)
```

# Decision Tree Model

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
Dt=DecisionTreeClassifier()
```
